```python
# data.yaml íŒŒì¼ì„ ìž‘ì„±í•˜ì—¬ ìˆ˜ì •ëœ ë°ì´í„°ì…‹ ê²½ë¡œì™€ í´ëž˜ìŠ¤ ì´ë¦„ì„ ì •ì˜í•©ë‹ˆë‹¤.
yaml_content = """
path: /content/yolo-V8
train: /content/yolo-V8/images/train
val: /content/yolo-V8/images/val

names:
  0: bike_stand
  1: bike_lie
  2: crosswalk
  3: braille_block
  4: bike_road
  5: car_road
"""

# yaml íŒŒì¼ ìƒì„±
with open('/content/yolo-V8/data.yaml', 'w') as f:
    f.write(yaml_content)
```

```python
from ultralytics import YOLO

# Download the pre-trained weights if they don't exist
!wget https://github.com/ultralytics/assets/releases/download/v0.0.0/yolov8m-seg.pt

# YOLOv8 ëª¨ë¸ ì´ˆê¸°í™”
model = YOLO('yolov8m-seg.pt')  # ë” í° YOLOv8 ëª¨ë¸ì˜ ì‚¬ì „ í•™ìŠµëœ ê°€ì¤‘ì¹˜ë¡œ ì´ˆê¸°í™”

# ëª¨ë¸ í•™ìŠµ ì‹œìž‘
model.train(
    data='/content/yolo-V8/data.yaml',   # ìˆ˜ì •ëœ data.yaml íŒŒì¼ ê²½ë¡œ
    epochs=100,                          # í•™ìŠµ ë°˜ë³µ íšŸìˆ˜
    imgsz=640,                           # ìž…ë ¥ ì´ë¯¸ì§€ í¬ê¸°
    batch=16,                            # ë°°ì¹˜ í¬ê¸°
    lr0=0.001,                           # ì´ˆê¸° í•™ìŠµë¥ 
    lrf=0.01,                            # ìµœì¢… í•™ìŠµë¥  ë¹„ìœ¨
    momentum=0.937,                      # ëª¨ë©˜í…€
    weight_decay=0.0005,                 # ê°€ì¤‘ì¹˜ ê°ì‡ 
    augment=True,                        # ë°ì´í„° ì¦ê°• í™œì„±í™”
    optimizer='Adam',                    # ì˜µí‹°ë§ˆì´ì € ì„¤ì • (SGD ë˜ëŠ” Adam)
    patience=15,                         # ì¡°ê¸° ì¢…ë£Œë¥¼ ìœ„í•œ ì¸ë‚´ ì—í­ ì„¤ì •
    save_period=5                        # ë§¤ 5 ì—í­ë§ˆë‹¤ ëª¨ë¸ ì €ìž¥
)
```

```python
# í•™ìŠµëœ ëª¨ë¸ì„ ì´ìš©í•´ í…ŒìŠ¤íŠ¸ ì´ë¯¸ì§€ì—ì„œ ê°ì²´ë¥¼ íƒì§€í•©ë‹ˆë‹¤.
# í…ŒìŠ¤íŠ¸í•  ì´ë¯¸ì§€ì˜ ê²½ë¡œë¥¼ ì§€ì •í•©ë‹ˆë‹¤.
test_image_path = '/content/yolo-V8/images/test'  # í…ŒìŠ¤íŠ¸í•  ì´ë¯¸ì§€ ê²½ë¡œ

# ì˜ˆì¸¡ ìˆ˜í–‰ ë° ê²°ê³¼ ì €ìž¥
results = model.predict(source=test_image_path, save=True)

# Since results is a list of Results objects, access the first element
# which corresponds to the prediction for the single image.
results = results[0]

# ì˜ˆì¸¡ ê²°ê³¼ ì¶œë ¥ -  'print' is not a valid method for the Results object.
# Use 'print(results)' or results.boxes.print() instead.
print(results) # Prints a summary of the results.
```
